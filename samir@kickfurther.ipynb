```python
import requests
```

```python
response = requests.get("https://api.openweathermap.org/data/2.5/onecall?lat=38.7267&lon=-9.1403&units=metric&appid=eb8abb514d4a3704ac0cbd7c1957f121")
print(response.status_code)
```

```python
print(response.json())
```

```python
import json

def jprint(obj):
    # create a formatted string of the Python JSON object
    text = json.dumps(obj, sort_keys=True, indent=4)
    print(text)

jprint(response.json())
```

```python
pass_times = response.json()['response']
jprint(pass_times)
```
